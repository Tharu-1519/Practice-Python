```python
#pandas is a global library function that is used for handling all related database functions
import pandas as pd
```

```python
#dataset is a variable which stores the file that we read using of pd .read-csv function
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#to display the "dataset varibale" 
dataset
```

```python
#create a new variable as independent for splitting the input dataset 
independent=dataset[["YearsExperience"]]
```

```python
#to display the input dataset - "independent"
independent
```

```python
#create another variable as dependent for splitting the output dataset 
dependent=dataset[["Salary"]]
```

```python
#to display the output dataset - "dependent"
dependent
```

```python
#split train dataset and test dataset 
#sklearn is another library setup for using the function -'train_test_split'
from sklearn.model_selection import train_test_split
#store the input dataset for both train and test as well as output dataset for both train and test in the 4 functions
x_train, x_test, y_train, y_test =train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#LinearRegression is a function used for the learning phase procedure
from sklearn.linear_model import LinearRegression
#regressor is a function which stores the function of LinearRegression
regressor=LinearRegression()
#fit method used to check the training dataset
regressor.fit(x_train,y_train)
```

```python
# to find the weight which is used for coef_ variable for calculating the weight
weight=regressor.coef_
weight
```

```python
# to find the bias used for intercept_ variable for calculating the bias/starting point/origin 
bias=regressor.intercept_
bias
```

```python
#Evalution Metrics
#predict the test dataset where we saved the model and we created a new variable for allocating memory.
y_pred=regressor.predict(x_test)
#r2 function call n python 
from sklearn.metrics import r2_score 
#call the y_test(actual test dataset) and y_pred(predicted dataset)
r_score=r2_score(y_test,y_pred)
r_score
#Note: "When the r_score value is near 1, then it is the good model"
```

```python
#How to save the model using python
import pickle
#name of the file that will be stored in the path
file_name="finalized_model_linear.sav"
```

```python
#Write binary means to write the file into the path 
pickle.dump(regressor,open("finalized_model_linear.sav",'wb'))
```

```python
#read binary and loaded _model variable 
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#passing the input to predict the data
result=loaded_model.predict([13]) 
```
